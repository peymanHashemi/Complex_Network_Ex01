```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
from numpy.linalg import matrix_power
```

```python
def Q2(N,k ,p_list):

  k_list = []
  ratio_list = []

  for p in p_list:
    graph_np = nx.erdos_renyi_graph(N, p, directed=False)
    edge_number =len(graph_np.edges())
    new_k = edge_number/N
    k_list.append(new_k)
    largest_cc = max(nx.connected_components(graph_np), key=len)
    size_cc = len(largest_cc)
    ratio = size_cc / N
    ratio_list.append(ratio)
    graph_np.clear
  return k_list , ratio_list
```

```python
N=[5,10,50,100,500,1000,5000]
for n in N:
    N = n
    k = [0,0.25,0.5,0.75,1,1.25,1.5,1.75,2,2.5,3,3.5,4,5,6,7]
    p_list = []
    for i in k:
        x = i/(N-1)
        p_list.append(x)

    k_list , ratio_list = Q2(N , k ,p_list )
    fig = plt.figure(figsize=(8,6))
    plt.xlabel('k')
    plt.ylabel('ratio')
    plt.title('N = {}'.format(N))
    plt.plot(k_list,ratio_list)
    plt.show()
    print('P = ',p_list)
```
